# Multi-Agent Conversation and Stand-up Comedy

```python
from autogen import ConversableAgent
import pprint

# ----------------------------
# GROQ API CONFIG
# ----------------------------
llm_config = {
    "config_list": [
        {
            "model": "llama-3.1-8b-instant",
            "api_key": "Your API KEY",   
            "base_url": "https://api.groq.com/openai/v1"
        }
    ]
}

# ----------------------------
#  AGENT 1: JOE
# ----------------------------
joe = ConversableAgent(
    name="Joe",
    system_message=(
        "You are Joe, a stand-up comedian. "
        "You tell short funny jokes and respond quickly."
    ),
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "I gotta go" in msg["content"]
)

# ----------------------------
# AGENT 2: CATHY
# ----------------------------
cathy = ConversableAgent(
    name="Cathy",
    system_message=(
        "You are Cathy, a stand-up comedian. "
        "You reply with witty punchlines and continue jokes."
    ),
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "I gotta go" in msg["content"]
)

# ----------------------------
# START CHAT
# ----------------------------
chat_result = joe.initiate_chat(
    recipient=cathy,
    message="I'm Joe. Cathy, let's do a funny stand-up comedy battle!",
    max_turns=4
)

# ----------------------------
# OUTPUTS
# ----------------------------
print("\n===== CHAT HISTORY =====\n")
pprint.pprint(chat_result.chat_history)

print("\n===== COST =====\n")
pprint.pprint(chat_result.cost)

print("\n===== SUMMARY =====\n")
pprint.pprint(chat_result.summary)
```
